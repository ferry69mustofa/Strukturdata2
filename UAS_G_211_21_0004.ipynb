```python
import matplotlib.pyplot as plt
import networkx as nx

print('Weighted-Graph Model :')
G = nx.Graph()
G.add_edge('A', 'B', weight=5)
G.add_edge('A', 'C', weight=7)
G.add_edge('A', 'D', weight=3)
G.add_edge('B', 'E', weight=2)
G.add_edge('B', 'F', weight=10)
G.add_edge('B', 'A', weight=5)
G.add_edge('C', 'A', weight=7)
G.add_edge('C', 'G', weight=1)
G.add_edge('D', 'A', weight=3)
G.add_edge('D', 'H', weight=11)
G.add_edge('E', 'B', weight=2)
G.add_edge('E', 'H', weight=9)
G.add_edge('F', 'B', weight=10)
G.add_edge('F', 'H', weight=4)
G.add_edge('G', 'C', weight=1)
G.add_edge('G', 'H', weight=6)
G.add_edge('H', 'F', weight=4)
G.add_edge('H', 'D', weight=11)
G.add_edge('H', 'E', weight=9)
G.add_edge('H', 'G', weight=6)

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos, node_size=800)
nx.draw_networkx_edges(G, pos, width=2)
nx.draw_networkx_labels(G, pos, font_size=20, font_family='sans-serif')

edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=12)

plt.axis('off')
plt.show()
```
